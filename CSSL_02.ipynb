##Take list of scores i.e 65, 70, 90, 85 and calculate avg

```python
cl = [65,70,90,85]
avg = sum(cl)/len(cl)
ma = max(cl)
mi = min (cl)
print(f'The average score is {avg} and maximum score is {ma} and minimum is {mi}')
print('The average score is %.1f and the maximum is %d and minimum is %d' %(avg,ma,mi))
```

#Funtions and if

```python
def student(score):
  '''This is a function that checks if a student has passed
  the pass mark is 40 '''
  if score>=40:
    #print('pass')
    return "pass"
  else:
    #print('fail')
    return "fail"

student(33)
```

##For loop

```python
for i in cl:
  print(f'i={i} {student(i)}')
```
